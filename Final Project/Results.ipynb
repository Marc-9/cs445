```python
import pandas as pd
```

```python
df = pd.read_csv('predictions_100.csv')
print(len(df))
df.head()
```

```python
def pick_Winner_on_Percent(df):
    matrix = [0,0,0,0,0,0,0,0,0]
    winners = 0
    cash = len(df)
    for i, row in df.iterrows():
        cash -= 1
        if row['My Win %'] >  row['My Draw %'] and row['My Win %'] > row['My Loss %']:
            if row['home_win'] == 1:
                winners += 1
                cash += 1 * row['home_odds']
                matrix[4] += 1
                continue
            else:
                if row['draw'] == 1:
                    matrix[1] += 1
                    continue
                if row['away_win'] == 1:
                    matrix[7] += 1
                    continue
        if row['My Draw %'] >  row['My Win %'] and row['My Draw %'] >  row['My Loss %']:
            if row['draw'] == 1:
                winners += 1
                cash += 1 * row['draw_odds']
                matrix[0] += 1
                continue
            else:
                if row['home_win'] == 1:
                    matrix[3] += 1               
                    continue
                if row['away_win'] == 1:
                    matrix[6] += 1
                    continue
        if row['My Loss %'] >  row['My Win %'] and row['My Loss %'] >  row['My Draw %']:
            if row['away_win'] == 1:
                cash += 1 * row['away_odds']
                winners += 1
                matrix[8] += 1
                continue
            else:
                if row['home_win'] == 1:
                    matrix[5] += 1               
                    continue
                if row['draw'] == 1:
                    matrix[2] += 1
                    continue
    print("Accuracy- " + str(winners/len(df)))
    print("Return = " + str(cash/len(df)))
    for x in matrix:
        print(x)
```

```python
pick_Winner_on_Percent(df)
```

```python
def pick_Winner_on_Odds(df):
    winners = 0
    cash = len(df)
    for i, row in df.iterrows():
        cash -= 1
        home_diff = row['home_odds'] - row['Home Odds']
        draw_diff = row['draw_odds'] - row['Draw Odds']
        away_diff = row['away_odds'] - row['Loss Odds']
        
        if home_diff > draw_diff and home_diff > away_diff:
            if row['home_win'] == 1:
                winners += 1
                cash += 1 * row['home_odds']
        if draw_diff > home_diff and draw_diff > away_diff:
            if row['draw'] == 1:
                winners += 1
                cash += 1 * row['draw_odds']
        if away_diff > home_diff and away_diff > draw_diff:
            if row['away_win'] == 1:
                cash += 1 * row['away_odds']
                winners += 1
    print("Accuracy- " + str(winners/len(df)))
    print("Return = " + str(cash/len(df)))
```

```python
pick_Winner_on_Odds(df)
```
