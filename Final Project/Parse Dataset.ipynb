```python
import os
import json
```

```python
unique_stats = set()
```

```python
for x in os.listdir('data'):
    if '.json' in x:
        with open('data/'+x, "r") as f:
            jd = json.loads(f.read())
            for stat in jd['liveData']['lineUp'][0]['stat']:
                unique_stats.add(stat['type'])
            for stat in jd['liveData']['lineUp'][1]['stat']:
                unique_stats.add(stat['type'])
```

```python
important = ['possLostAll', 'shotFastbreak', 'lastManTackle', 'totalFinalThirdPasses'
            , 'putThrough', 'accurateGoalKicks', 'aerialLost', 'totalLayoffs', 'interception', 'interceptionsInBox',
            'totalLaunches', 'errorLeadToShot', 'effectiveHeadClearance', 'sixYardBlock', 'interceptionWon',
            'attemptedTackleFoul', 'crosses18yardplus', 'attemptsConcededIbox', 'attFastbreak',
            'finalThirdEntries', 'possWonAtt3rd', 'attemptsObox', 'duelLost', 'wonContest', 'accurateFreekickCross',
            'openPlayPass', 'possWonMid3rd', 'totalFwdZonePass', 'totalPass', 'accuratePass',
            'saves', 'possWonDef3rd', 'attemptsIbox', 'touches', 'accurateLaunches', 'headClearance',
            'effectiveBlockedCross', 'attOpenplay', 'fwdPass', 'backwardPass', 'totalFastbreak',
            'touchesInOppBox', 'shotOffTarget', 'attCorner', 'blockedPass', 'accurateBackZonePass',
            'possessionPercentage', 'totalCornersIntobox', 'totalThroughBall', 'overrun', 'wonTackle',
            'totalClearance', 'attSetpiece', 'penAreaEntries', 'ballRecovery', 'cornerTaken', 'cornerTaken',
            'fouledFinalThird', 'dispossessed']
```

```python
with open('parsed.csv', "w+") as f2:
    for x in os.listdir('data'):
        if '.json' in x:
            with open('data/'+x, "r") as f:
                jd = json.loads(f.read())
                home_team = jd['matchInfo']['contestant'][0]['name']
                away_team = jd['matchInfo']['contestant'][1]['name']
                date = jd['matchInfo']['date'][0:-1]
                entry = [home_team,away_team,date]
                if(jd['liveData']['matchDetails']['winner'] == 'away'):
                    entry.append(0)
                    entry.append(1)
                    entry.append(0)
                if(jd['liveData']['matchDetails']['winner'] == 'draw'):
                    entry.append(0)
                    entry.append(0)
                    entry.append(1)
                if(jd['liveData']['matchDetails']['winner'] == 'home'):
                    entry.append(1)
                    entry.append(0)
                    entry.append(0)
                for x in important:
                    not_found = True
                    for stat in jd['liveData']['lineUp'][0]['stat']:
                        if stat['type'] == x:
                            entry.append(stat['value'])
                            not_found = False
                            break
                    if not_found:
                        entry.append(0)
                    not_found = True
                    for stat in jd['liveData']['lineUp'][1]['stat']:
                        if stat['type'] == x:
                            entry.append(stat['value'])
                            not_found = False
                            break
                    if not_found:
                        entry.append(0)
                f2.write(','.join(map(str,entry)))
                f2.write("\n")
```

```python
st = ""
for x in important:
    st += 'home_' + x + ',' + 'away_' + x +','
```
